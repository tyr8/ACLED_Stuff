## Libraries

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize 
```

## Info of counts on by ACLED region and total count

### Overall Counts

```python
##Make request
URL = "https://api.acleddata.com/region/read?terms=accept"
r = requests.get(url = URL)
data = r.json()
```

```python
##Normalize Json
df = json_normalize(data, 'data')
df
```

```python
df['count'] = df['event_count'].astype(int)
total = df['count'].sum(axis=0)
print(total)
```

### Pulling event data points by data range

```python
##Variables
start_date = '2019-01-01'
end_date = '2020-03-14'
lim = '750000'
spec_reg = 'No'
region = '15'
```

```python
##Building of Parameters  - Should not need to change
URL = 'https://api.acleddata.com/acled/read?terms=accept'
if spec_reg == 'No':
    PARAMS = "&limit="+lim+"&event_date="+start_date+"|"+end_date+"&event_date_where=BETWEEN"
elif spec_reg == 'Yes':
    PARAMS = "&limit="+lim+"&region="+region+"&event_date="+start_date+"|"+end_date+"&event_date_where=BETWEEN"
```

```python
##Request for data
r = requests.get(url = URL, params = PARAMS)
data = r.json()
```

```python
df = json_normalize(data, 'data')
df
```
