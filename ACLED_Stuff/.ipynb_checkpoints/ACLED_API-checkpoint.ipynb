## Libraries

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize 
```

## Info on ACLED Regions

### Overall Counts

```python
##ACLED Region URL 
URL = "https://api.acleddata.com/region/read?terms=accept"
```

```python
##Make request
r = requests.get(url = URL)
data = r.json()

```

```python
##Normalize Json
df = json_normalize(data, 'data')
print(df)
```

### Overall Count by Time Range 

```python
# ACLED Region URL 
URL = "https://api.acleddata.com/acled/read?terms=accept"
PARAMS = "&first_event_date>={2019-01-01}&last_event_date<={2019-01-03}"
```

```python
r = requests.get(url = URL, params = PARAMS)
data = r.json()
```

```python
df = json_normalize(data, 'data')
print(df)
```

### Pulling event data points

```python
URL = "https://api.acleddata.com/acled/read?terms=accept"
PARAMS = '&year=2019&limit=100'
r = requests.get(url = URL, params = PARAMS )
data = r.json()
print(data)
```

```python
df = json_normalize(data, 'data')
print(df)
```
